# 7. Menampilkan Hasil Prediksi

```python
from PIL import Image
import matplotlib.pyplot as plt

# List of image paths
image_paths = [
    ('/content/test/images/data2_6_jpg.rf.bbd5d1261e1d63b4dee52413f7099347.jpg',
     '/content/runs/detect/predict2/data2_6_jpg.rf.bbd5d1261e1d63b4dee52413f7099347.jpg'),
    ('/content/test/images/WIN_20240512_07_42_44_Pro-2-_jpg.rf.1568fe4c4ffe99356838394c6a076c67.jpg',
     '/content/runs/detect/predict4/WIN_20240512_07_42_44_Pro-2-_jpg.rf.1568fe4c4ffe99356838394c6a076c67.jpg'),
    ('/content/test/images/WIN_20240629_11_30_10_Pro_jpg.rf.06605cba68d89cd0f82962f1d64c876a.jpg',
     '/content/runs/detect/predict5/WIN_20240629_11_30_10_Pro_jpg.rf.06605cba68d89cd0f82962f1d64c876a.jpg'),
    ('/content/test/images/20240311_551_jpg.rf.0e57b71fbe8086004f5db9786936eb15.jpg',
     '/content/runs/detect/predict7/20240311_551_jpg.rf.0e57b71fbe8086004f5db9786936eb15.jpg'),
    ('/content/train/images/20240311_809_jpg.rf.e74fc5a3d3af756e1c22529aba7ce803.jpg',
     '/content/runs/detect/predict8/20240311_809_jpg.rf.e74fc5a3d3af756e1c22529aba7ce803.jpg'),
    # Add more image path pairs as needed
]

# Number of comparisons
num_comparisons = len(image_paths)

# Create a figure with subplots
fig, axes = plt.subplots(num_comparisons, 2, figsize=(8, 4 * num_comparisons))

for i, (initial_path, predicted_path) in enumerate(image_paths):
    # Open images using PIL
    initial_image = Image.open(initial_path)
    predicted_image = Image.open(predicted_path)

    # Display the initial image
    axes[i, 0].imshow(initial_image)
    axes[i, 0].axis('off')  # Turn off axis labels
    axes[i, 0].set_title(f'Image Awal {i+1}')

    # Display the predicted image
    axes[i, 1].imshow(predicted_image)
    axes[i, 1].axis('off')  # Turn off axis labels
    axes[i, 1].set_title(f'Image Predict {i+1}')

# Adjust layout
plt.tight_layout()
# Show the plot
plt.show()
```
